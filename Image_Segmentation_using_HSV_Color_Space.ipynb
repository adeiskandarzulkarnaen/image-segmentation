Import library yang dibutuhkan

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import io, measure
from scipy import ndimage as nd
```

Muat gambar dan tampilkan (dalam format RGB). Pastikan mengupload gambar pingpong.jpeg terlebih dahulu ke dalam folder `/content/`

```python
img = io.imread('/content/pingpong.jpeg')
plt.imshow(img)
```

Konversi ke dalam format HSV

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

Buat fungsi segmentasi gambar

```python
def segmentation_image(image, lower_bound, upper_bound):
  # segmentation the image
  segmented_image = cv2.inRange(image, lower_bound, upper_bound)

  # image morphology
  transformed_image = nd.binary_closing(segmented_image, np.ones((50, 3)))
  return transformed_image
```

Segmentasi bola orange

```python
min_mask = (5, 150, 0)
max_mask = (30, 255, 255)

segmented_image = segmentation_image(img_hsv, min_mask, max_mask)

# Show original image
plt.imshow(img)
plt.show()
# Show segmented image
plt.imshow(segmented_image)
```

Segmentasi bola pink

```python
min_mask = (140, 80, 0)
max_mask = (170, 255, 255)

segmented_image = segmentation_image(img_hsv, min_mask, max_mask)

# Show original image
plt.imshow(img)
plt.show()
# Show segmented image
plt.imshow(segmented_image)
```

Segmentasi bola hijau

```python
min_mask = (40, 100, 0)
max_mask = (65, 255, 255)

segmented_image = segmentation_image(img_hsv, min_mask, max_mask)

# Show original image
plt.imshow(img)
plt.show()
# Show segmented image
plt.imshow(segmented_image)
```
